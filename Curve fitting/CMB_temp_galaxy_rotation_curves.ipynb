## Obtaining CMB Temperature from COBE satellite data

```python
data = ascii.read('COBE_CMB_data.txt')
freq = data['frequency (Hz)']
brightness = data['brightness (ergs/s/cm^2/sr/Hz)']

print(freq.dtype, freq.shape, brightness.dtype, brightness.shape)
```

```python
# defining data in cgs: ergs, cm, sec
h = 6.626 * 10**(-27) # erg s
k = 1.38 * 10**(-16) # erg K^-1
c_incms = 2.997924 * 10**(10) # cm/s

k1 = 2*h/c_incms**2
k2 = h / k

print(k1, k2)
```

```python
fig = plt.figure(figsize=(15, 5))
fig.suptitle('plots of B_lambda(T) vs Frequency')

ax1 = plt.subplot(1, 2, 1)
plt.plot(freq, brightness, color='black')
ax1.set_title('Normal')
plt.ylabel('B_lambda')
plt.xlabel('frequency')

ax2 = plt.subplot(1, 2, 2)
plt.scatter(freq, brightness, s=15, c='black')
plt.yscale('log')
plt.xscale('log')
ax2.set_title('log-log')
plt.ylabel('B_lambda')
plt.xlabel('frequency')


plt.savefig('b_lambda.jpg')
```

```python
def planck_func(nu, T):
    B_nu = k1 * (nu**3) * ( (np.exp(k2 * nu / T) - 1)**(-1) )
    return B_nu
```

```python
param, param_cov = curve_fit(planck_func, freq, brightness)
```

```python
print('CMB Temp =', f'{param[0]:.4f}', 'K')
print('fit covariance = ', f'{param_cov[0,0]:.2e}')
```

## Plotting Galaxy rotation curves using Hydrogen-21cm lines

```python
nu_21cm = 1420.40575e+6
# lambda_21cm = c/nu_21cm # acutually you should use this value for 21cm line wavelength
lambda_21cm = 0.21 # in m, but the guy who generated the data assumed this 
```

```python
def gaussian(x, mu, sigma):
    y = ((sigma * np.sqrt(2* np.pi) )**(-1)) * np.exp(-0.5*((x-mu)/sigma)**2)
    return y
```

```python
def get_radial_velocity(filepath, redshift_list, velocity_list):
    line_data = ascii.read(filepath)
    lambdas = line_data['wavelngth (cm)']
    brightnesses = line_data['brightness']
    
    line_params, covariance = curve_fit(gaussian, lambdas, brightnesses, p0=(10, 20))
    central_lambda = 0.01 * line_params[0] # mean of the gaussian, refers to central wavelength (in meters)
    
#     get redshifts
    z = (central_lambda - lambda_21cm) / lambda_21cm
    
#     central_f = c / central_lambda
#     z = (nu_21cm - central_f) / central_f
    redshift_list.append(z)
    
    # get velocities
    v = c * ((z+1)**2 - 1) / ((z+1)**2 + 1)
    velocity_list.append(v)
```

```python
distance_list = []
redshift_list = []
velocity_list = []

for entity in os.scandir('./galaxy_21cm_spectrum/'):
    if entity.is_file():
        path = entity.path
        
        # get distances
        d = re.findall("\d+\.\d+", path)
        distance_list.append(float(d[0]))
        
        get_radial_velocity(path, redshift_list, velocity_list)

print(distance_list)
print(redshift_list)
print(velocity_list)
```

```python
plt.scatter(distance_list, velocity_list, [10]*len(distance_list), color='blue', marker='o',)
plt.ylabel('Velocity (km/s)')
plt.xlabel('Radial Distance (kpc)')
plt.savefig('GalRotcurve2.jpg')
```
