```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii

from scipy.optimize import curve_fit
import emcee
import corner
```

```python
data = ascii.read('./tauceti.caltech.edu_kunal_gw170817_gw170817_afterglow_data_full.txt')
data
```

```python
data_by_telescope = data.group_by('Telescope')
keys = data_by_telescope.groups.keys
print(keys)
```

```python
vla_mask = data_by_telescope.groups.keys['Telescope'] == 'VLA'
vla_data = data_by_telescope.groups[vla_mask]
vla_freqs = vla_data.group_by('Freq')
vla_3GHz_mask = vla_freqs.groups.keys['Freq'] == 3e+9
vla_3GHz_data = vla_freqs.groups[vla_3GHz_mask]


chandra_mask = data_by_telescope.groups.keys['Telescope'] == 'Chandra'
chandra_data = data_by_telescope.groups[chandra_mask]

print(chandra_data)
```

```python
# vla 3GHz data extracts

# F_nu
vla_F_nu = np.array(vla_3GHz_data['FluxD'])
vla_F_nu = np.array([i.replace('<', '') for i in vla_F_nu]).astype(float)

# t
vla_t = np.array(vla_3GHz_data['T'])

# nu
vla_nu = np.array(vla_3GHz_data['Freq'])

# errors
vla_errors = np.array(vla_3GHz_data['FluxDErr'])

print(vla_nu)
```

```python
fig = plt.figure(figsize=(10,4))
ax = plt.axes()
ax.errorbar(vla_t[2:], vla_F_nu[2:], vla_errors[2:], ecolor='red', marker='s', mfc='black')
ax.set_xscale('log')
ax.set_yscale('log')
```

```python
# chandra data extracts
chandra_F_nu = np.array(chandra_data['FluxD'])
chandra_F_nu = np.array([i.replace('<', '') for i in chandra_F_nu]).astype(float)

# t
chandra_t = np.array(chandra_data['T'])

# nu
chandra_nu = np.array(chandra_data['Freq'])

# errors
chandra_errors = np.array(chandra_data['FluxDErr'])

print(chandra_nu)
```

```python
# remove first 2 vla entries and first chandra entry (outliers) and compile data

vla_F_nu = vla_F_nu[2:]
chandra_F_nu = chandra_F_nu[1:]

vla_t = vla_t[2:]
chandra_t = chandra_t[1:]

vla_nu = vla_nu[2:]
chandra_nu = chandra_nu[1:]

vla_errors = vla_errors[2:]
chandra_errors = chandra_errors[1:]

F_nu = np.concatenate((vla_F_nu, chandra_F_nu))
t = np.concatenate((vla_t, chandra_t))
nu = np.concatenate((vla_nu, chandra_nu))
errors = np.concatenate((vla_errors, chandra_errors))
print(F_nu.shape, t.shape, nu.shape, errors.shape, sep='\n')
```

```python
# model to be fit using mcmc
def smoothBrokenPowerLaw(parameters, x):
    nu, t = x
    beta, Fp, tp, s, alpha1, alpha2 = parameters
    
    term1 = (2**(1/s)) * ( (nu / 3e+9)**beta ) * Fp
    term2 = (t / tp)**(-s*alpha1)
    term3 = (t / tp)**(-s*alpha2)
    term4 = (term2 + term3) ** (-1/s)
    
    return term1 * term4

# likelihood function
def lnlike(parameters, x, y, yerr):
    return -0.5 * np.sum( ( (y - smoothBrokenPowerLaw(parameters, x)) /yerr ) ** 2)

# function to check/set prior values to this bayesian mcmc sampler
def lnprior(parameters):
    beta, Fp, tp, s, alpha1, alpha2 = parameters
    if (-1 < beta < 0) and (80 < Fp < 120) and (70 < tp < 200) and (1 < s < 15) and (0 < alpha1 < 2) and (-2 < alpha2 < 0):
        return 0.0
    return -np.inf

def lnprob(parameters, x, y, yerr):
    lp = lnprior(parameters)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike(parameters, x, y, yerr)

def samplerFunction(p0,num_walkers,num_iterations,n_dim,lnprob,data):
    sampler = emcee.EnsembleSampler(num_walkers, n_dim, lnprob, args=data)

    # print("Running burn-in...")
    p0, _, _ = sampler.run_mcmc(p0, 100)
    sampler.reset()

    # print("Running production...")
    pos, prob, state = sampler.run_mcmc(p0, num_iterations)

    return sampler, pos, prob, state
```

```python
def plotter(sampler,x,y):
    plt.ion()
    plt.plot(x, y,label='Brightness vs Wavelength')
    samples = sampler.flatchain
    for parameters in samples[np.random.randint(len(samples), size=100)]:
        plt.plot(x, smoothBrokenPowerLaw(parameters, x), color="r", alpha=0.1)
    plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
    plt.xlabel('Wavelength')
    plt.ylabel('Brightness')
    plt.legend()
    plt.show()
```

```python
def MCMCFitParams(x, y, y_error, num_walkers, num_iterations, initial_params):
    # setting errors in y-values (brightness)
#     y_error = error_percent * 0.01 * np.mean(y)

    # package data as required by emcee sampler
    data = (x, y, y_error)

    # give initial values
    n_dim = len(initial_params)
    p0 = [np.array(initial_params) + 1e-7 * np.random.randn(n_dim) for i in range(num_walkers)]
    
    sampler, pos, prob, state = samplerFunction(p0,num_walkers,num_iterations,n_dim,lnprob,data)

    samples = sampler.flatchain
    return samples[np.argmax(sampler.flatlnprobability)]
```

```python
initial = np.array([-0.7, 90, 100, 2, 1, -1])
x = (nu, t)
y = F_nu
y_err = errors
par = MCMCFitParams(x, y, y_err, 250, 10000, initial)
print(par)
```

```python
plt.plot(vla_t, smoothBrokenPowerLaw(par, (vla_nu, vla_t)), color='black')
plt.plot(chandra_t, smoothBrokenPowerLaw(par, (chandra_nu, chandra_t)), color='black', label='_nolegend_')
# plt.scatter(x[1], y)
plt.errorbar(vla_t, vla_F_nu, vla_errors, ecolor='red', marker='.', linestyle='', markersize=5, c='blue')
plt.errorbar(chandra_t, chandra_F_nu, chandra_errors, ecolor='red', marker='.', linestyle='', markersize=5, c='black')
plt.xscale('log')
plt.yscale('log')
plt.ylabel(r'Flux Density $F_\nu$ ($\mu \,\, Jy$)')
plt.xlabel('Time since merger t (s)')
plt.legend(['Curve fit based on SBPL', 'VLA', 'Chandra'])

plt.savefig('light_curve_fitted', bbox_inches='tight')
plt.show()
```

```python
# getting posterior spread between parameters


# setting errors in y-values (brightness)
y_err2 = vla_errors

# nums
num_walkers_test = 250
num_iterations_test = 10000

# package data as required by emcee sampler
data_test = (x, y, y_err)

# give initial values
initial_test = np.array([-0.7, 90, 100, 2, 1, -1])
n_dim_test = len(initial_test)
p0 = [np.array(initial_test) + 1e-4 * np.random.randn(n_dim_test) for i in range(num_walkers_test)]


sampler, pos, prob, state = samplerFunction(p0,num_walkers_test,num_iterations_test,n_dim_test,lnprob,data_test)

samples = sampler.flatchain

labels = [r'$\beta$', r'$F_p$', r'$t_p$', 's', r'$\alpha 1$', r'$\alpha 2$']
fig = corner.corner(samples,show_titles=True,labels=labels,plot_datapoints=True,quantiles=[0.16, 0.5, 0.84])
```
