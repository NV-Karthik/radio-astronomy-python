# Galaxy Rotation Curve from Brightness Data of H-21cm Line

The redshift data of Hydrogen-21cm line reveals the radial velocities of each star. This is used to plot the Galaxy Rotation Curve. The redshift data is obtained by fitting a gaussian to the data points, the difference between the central peak wavelength and the original H-21cm line gives us the redshift.

The gaussian parameters $\mu$ and $\sigma$ are sampled using Markov Chain Monte Carlo (MCMC) method

```python
import numpy as np
import matplotlib.pyplot as plt
import emcee
import corner
from astropy.io import ascii 

import os
import re

# import tikzplotlib
```

```python
# import one file data to examine and plot
data = ascii.read('./galaxy_21cm_spectrum/spectrum_d_0.69_kpc.txt')
data
```

```python
plt.plot(data['wavelngth (cm)'], data['brightness'], color='blue')
plt.ylabel('Brightness')
plt.xlabel('Wavelength (cm)')
plt.savefig('init_waveform.jpg')
```

```python
brightness = np.array(data['brightness'])
wavelength = np.array(data['wavelngth (cm)'])

print(brightness.shape, wavelength.shape)
```

```python
# mcmc fit model
def gaussian(parameter, x):
    amp, mu, sigma, y0 = parameter # unpacking the parameter vector
    y = (amp * np.exp(-0.5*((x-mu)/sigma)**2)) + y0
    return y

def gaussian2(parameter, x):
    amp, mu, sigma = parameter # unpacking the parameter vector
    a = amp / ((sigma) * (2*np.pi)**0.5)
    y = a * np.exp(-0.5*((x-mu)/sigma)**2)
    return y

# likelihood function
def lnlike(parameter, x, y, yerr):
    return -0.5 * np.sum( ( (y - gaussian2(parameter, x)) /yerr ) ** 2)

# function to check/set prior values to this bayesian mcmc sampler
def lnprior(parameter):
    amp, mu, sigma, y0 = parameter
    if (4000 < amp < 7000) and (20.50 < mu < 21.50) and (-0.5 < sigma < 3) and (-0.5 < y0 < 5):
        return 0.0
    return -np.inf

def lnprior2(parameter):
    amp, mu, sigma = parameter
    if (20.50 < mu < 21.50) and (-0.5 < sigma < 1) and (amp > 0):
        return 0.0
    return -np.inf

def lnprob(parameter, x, y, yerr):
    lp = lnprior(parameter)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike(parameter, x, y, yerr)

def lnprob2(parameter, x, y, yerr):
    lp = lnprior2(parameter)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike(parameter, x, y, yerr)

def samplerFunction(p0,num_walkers,num_iterations,n_dim,lnprob,data):
    sampler = emcee.EnsembleSampler(num_walkers, n_dim, lnprob, args=data)

    # print("Running burn-in...")
    p0, _, _ = sampler.run_mcmc(p0, 100)
    sampler.reset()

    # print("Running production...")
    pos, prob, state = sampler.run_mcmc(p0, num_iterations)

    return sampler, pos, prob, state

def samplerFunction2(p0,num_walkers,num_iterations,n_dim,lnprob2,data):
    sampler = emcee.EnsembleSampler(num_walkers, n_dim, lnprob2, args=data)

    # print("Running burn-in...")
    p0, _, _ = sampler.run_mcmc(p0, 100)
    sampler.reset()

    # print("Running production...")
    pos, prob, state = sampler.run_mcmc(p0, num_iterations)

    return sampler, pos, prob, state
```

```python
def plotter(sampler,x,y):
    plt.ion()
    plt.plot(wavelength,brightness,label='Brightness vs Wavelength')
    samples = sampler.flatchain
    for parameter in samples[np.random.randint(len(samples), size=100)]:
        plt.plot(wavelength, gaussian(parameter, wavelength), color="r", alpha=0.1)
    plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
    plt.xlabel('Wavelength')
    plt.ylabel('Brightness')
    plt.legend()
    plt.show()
```

```python
def MCMCFitParams(x, y, error_percent, num_walkers, num_iterations, initial_params):
    # setting errors in y-values (brightness)
    y_error = error_percent * 0.05 * np.mean(y)

    # package data as required by emcee sampler
    data = (x, y, y_error)

    # give initial values
    n_dim = len(initial_params)
    p0 = [np.array(initial_params) + 1e-7 * np.random.randn(n_dim) for i in range(num_walkers)]
    
    sampler, pos, prob, state = samplerFunction(p0,num_walkers,num_iterations,n_dim,lnprob,data)

    samples = sampler.flatchain
    return samples[np.argmax(sampler.flatlnprobability)]

def MCMCFitParams2(x, y, error_percent, num_walkers, num_iterations, initial_params):
    # setting errors in y-values (brightness)
    y_error = error_percent * 0.01 * np.mean(y)

    # package data as required by emcee sampler
    data = (x, y, y_error)

    # give initial values
    n_dim = len(initial_params)
    p0 = [np.array(initial_params) + 1e-7 * np.random.randn(n_dim) for i in range(num_walkers)]
    
    sampler, pos, prob, state = samplerFunction2(p0,num_walkers,num_iterations,n_dim,lnprob2,data)

    samples = sampler.flatchain
    return samples[np.argmax(sampler.flatlnprobability)]
```

```python
initial = np.array([5000, 21, 1, 0.1])
initial2 = np.array([100, 21, 0.5])
par = MCMCFitParams2(wavelength, brightness, 5, 256, 10000, initial2)
print(par[0], par[1], par[2])
```

```python
plt.plot(wavelength, gaussian2(par, wavelength), color='b')
plt.scatter(wavelength, brightness, color='r')
plt.xlabel('wavelength (cm)')
plt.ylabel('brightness')
plt.savefig('fitted.jpg')
```

```python
# getting posterior spread between parameters


# setting errors in y-values (brightness)
brightness_error = 5 * 0.01 * np.mean(brightness)

# nums
num_walkers_test = 500
num_iterations_test = 10000

# package data as required by emcee sampler
data_test = (wavelength, brightness, brightness_error)

# give initial values
initial_test = np.array([15, 20.60, 0.25]) 
n_dim_test = len(initial_test)
p0 = [np.array(initial_test) + 1e-7 * np.random.randn(n_dim_test) for i in range(num_walkers_test)]


sampler, pos, prob, state = samplerFunction2(p0,num_walkers_test,num_iterations_test,n_dim_test,lnprob2,data_test)

samples = sampler.flatchain

labels = ['amp', 'mu', 'sigma']
fig = corner.corner(samples,show_titles=True,labels=labels,plot_datapoints=True,quantiles=[0.16, 0.5, 0.84])
```

```python
# parameters to get radial velocities

nu_21cm = 1420.40575e+6 # also correct
# lambda_21cm = c/nu_21cm # correct value for 21cm line wavelength
lambda_21cm = 0.21 # in m, wrong but assume this for correct rotation curve
c = 299792458 # in m
```

```python
# to get redshift and radial velocities from brightness data
def get_redshift_velocity(filepath):
    line_data = ascii.read(filepath)
    wavelength_data = line_data['wavelngth (cm)']
    brightness_data = line_data['brightness']
    
    # using scipy's curvefit function
#     line_params, covariance = curve_fit(gaussian, lambdas, brightnesses, p0=(10, 20))
#     central_lambda = 0.01 * line_params[0] # mean of the gaussian, refers to central wavelength (in meters)
    
    # using MCMC to sample parameters (curve fit)
    initial_parameters = np.array([15, 21, 0]) # pass in initial parameters for fit, mu and sigma
    line_params_mcmc = MCMCFitParams2(wavelength_data, brightness_data, 1, 256, 1000, initial_parameters)
    central_lambda = 0.01 * line_params_mcmc[1]
    
    # get redshifts
    z = (central_lambda - lambda_21cm) / lambda_21cm
    
    # central_f = c / central_lambda
    # z = (nu_21cm - central_f) / central_f
    
    # get velocities
    v = c * ((z+1)**2 - 1) / ((z+1)**2 + 1)
    
    return z, v
```

```python
distance_list = []
redshift_list = []
velocity_list = []

for entity in os.scandir('./galaxy_21cm_spectrum/'):
    if entity.is_file():
        path = entity.path
        
        # get distances
        d = re.findall("\d+\.\d+", path)
        distance_list.append(float(d[0]))
        
        z, v = get_redshift_velocity(path)
        redshift_list.append(z)
        velocity_list.append(v)

print(distance_list)
print(redshift_list)
print(velocity_list)
```

```python
plt.scatter(distance_list, velocity_list, color='black', marker='.', s=80)
plt.ylabel('Velocity (km/s)')
plt.xlabel('Radial Distance (kpc)')
plt.savefig('grc_result_plot.png', bbox_inches='tight')
plt.show()
```
